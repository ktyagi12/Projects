```python
malnourish_param = ['AGE','HEIGHT','WEIGHT','BMI','FAT_CONTAINS']
```

```python
X = df[malnourish_param]
Y = df.MALNOURISHED
print (X)
print ('\n\n',Y)
```

```python
from sklearn.cross_validation import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X, Y ,test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()

logReg.fit(xtrain, ytrain)
```

```python
y_pred = logReg.predict(xtest)
```

```python
y_pred
```

```python
ytest
```

```python
from sklearn import metrics
print (round(metrics.accuracy_score(ytest, y_pred) * 100))
```

```python
print ('Wrongly predicted rows count: ')
print ((ytest != y_pred) .sum())
```

```python
print ('Correctly predicted rows count: ')
print ((ytest == y_pred) .sum())
```

```python
conf = metrics.confusion_matrix(ytest,y_pred)
print (conf)
```

## True:  Correctly predicted

## False: Wrongly predicted

## Positive: has malnutrition

## Negative: has not malnutrition


### TP: We predict right that he has malnutrition
### TN: We predict right that he does not have malnutrition
### FP: We predict wrong that he has malnutrition
### FN: We predict wrong that he does not have malnutrition


   0     1
0  TP   FN

1  FP   TN

```python
true_val = conf[0,0]  +conf[1,1]
true_val

accuracy = true_val / conf.sum()
accuracy
```

### ACCURACY: (TP + TN) / Total No of Observation


### CLASSIFICATION ERROR:  1- ACCURACY
    
    OR
    
### It tells how much wrong you have predicted


    OR
    
### (FP + FN) / Total

```python
false_values = 1 - accuracy


class_err = (conf[0,1] + conf[1,0])/ conf.sum()
```
