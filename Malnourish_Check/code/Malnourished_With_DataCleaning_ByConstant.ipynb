```python
l1 = df.columns
l1
```

```python
for i in  l1:
    if(i == 'AGE'):
        df[i] = df[i].fillna(0)
    #elif(i == 'EMAIL'):
     #   df[i] = df[i].fillna(desc_data.loc['mean',i])
    #elif(i == 'ADDRESS'):
     #   df[i] = df[i].fillna(desc_data.loc['mean',i])
    elif(i == 'CONTACT'):
        df[i] = df[i].fillna(0)
    elif(i == 'HEIGHT'):
        df[i] = df[i].fillna(0)
    elif(i == 'WEIGHT'):
        df[i] = df[i].fillna(0)
    elif(i == 'BMI'):
        df[i] = df[i].fillna(0)
    elif(i == 'FAT_CONTAINS'):
        df[i] = df[i].fillna(0)       
    
```

```python
df
```

```python
malnourish_param = ['AGE','HEIGHT','WEIGHT','BMI','FAT_CONTAINS']
```

```python
X = df[malnourish_param]
Y = df.MALNOURISHED
print (X)
print ('\n\n',Y)
```

```python
from sklearn.cross_validation import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X, Y ,test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()

logReg.fit(xtrain, ytrain)
```

```python
y_pred = logReg.predict(xtest)
```

```python
y_pred
```

```python
ytest
```

```python
from sklearn import metrics
print (round(metrics.accuracy_score(ytest, y_pred) * 100))
```
