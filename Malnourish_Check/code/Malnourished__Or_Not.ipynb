```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
path = r"D:\KTyagi\KT\Python_Workspace\DataScience\Malnourishment_Prediction1.xlsx"
df = pd.read_excel(path)
df
```

```python
malnourish_param = ['AGE','HEIGHT','WEIGHT','BMI','FAT_CONTAINS']
```

```python
X = df[malnourish_param]
Y = df.MALNOURISHED
print (X)
print ('\n\n',Y)
```

```python
from sklearn.cross_validation import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X, Y ,test_size = 0.25)
```

```python
logReg = LogisticRegression()

logReg.fit(xtrain, ytrain)
```

```python
y_predicted = logReg.predict(xtest)
```

```python
print ('Predicted Values: \t',y_predicted)
```

```python
print ('Actual Values:\n',ytest)
```

```python
from sklearn import metrics
print (round(metrics.accuracy_score(ytest, y_predicted) * 100))
```
