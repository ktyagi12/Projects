```python
import pandas as pd
from sklearn import linear_model as lm
import numpy as np
import matplotlib.pyplot as plt

path = r"TimeWorkAnalysis.xlsx"
df = pd.read_excel(path)
print ('Data: \n', df)

A_days = df['A']
B_days = df['B']
A_B_days = df['A_B_C_D']
```

plt.scatter(A_days, A_B_days, color = 'r', marker = 'o', s = 40)
plt.title('A_days Vs A_B_days')
plt.xlabel('A_days')
plt.ylabel('A_B_days')
plt.show()


plt.scatter(B_days, A_B_days, color = 'r', marker = 'o', s = 40)
plt.title('B_days Vs A_B_days')
plt.xlabel('B_days')
plt.ylabel('A_B_days')
plt.show()

```python
X = df[['A','B','C','D']]
Y = df['A_B_C_D']
reg = lm.LinearRegression()     # reg is the obj of class LinearRegression
reg.fit(X,Y)            # Analyses the relation b/w X and Y
print ('a = \n ',reg.intercept_)
print ('b = \n ',reg.coef_)
```

```python
print("predicted Work:",reg.predict([[4, 2,1,6]]))
```
